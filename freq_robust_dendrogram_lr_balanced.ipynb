```python
import pandas as pd
import numpy as np
import random
```

```python
seed = 42
```

```python
np.random.seed(seed)
random.seed(seed)
```

```python
# 예제 데이터셋 로드
X = pd.read_csv("./data/train.csv")
y = pd.read_csv("./data/label_encoded_y.csv")
X_test = pd.read_csv("./data/test.csv")
```

```python
X = X.drop(["ID", "SUBCLASS"], axis = 1)
```

```python
X.shape
```

```python
X_encoded.describe()
```

```python
X["A2M"].unique()
```

```python
# 변환 함수 정의
def mutation_count(value):
    if pd.isna(value) or value == 'WT':
        return 0
    else:
        return len(value.split())
```

```python
# 각 셀의 값을 변환
X_encoded = X.map(mutation_count)
```

```python
X_test = X_test.drop(['ID'], axis = 1)
```

```python
# 각 셀의 값이 float형인지 확인
for column in X_test.columns:
    for value in X_test[column]:
        if isinstance(value, float):
            print(f"{column}의 {value}는 float형입니다.")
            break
```

```python
X_test["AK2"].unique()
```

```python
X_test_encoded = X_test.map(mutation_count)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_val, y_train, y_val = train_test_split(X_encoded, y, stratify=y, 
                                                  test_size=0.2, random_state=seed)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)
X_test_scaled = scaler.transform(X_test_encoded)
```

```python
import scipy.cluster.hierarchy as sch
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
X_feature_wise = X_train_scaled.T
corr_matrix = np.corrcoef(X_feature_wise)
```

```python
corr_matrix = np.nan_to_num(corr_matrix, nan=0.0, posinf=1.0, neginf=-1.0)
linked = sch.linkage(corr_matrix, method='average')
```

```python
plt.figure(figsize=(10, 7))
dendrogram = sch.dendrogram(linked, labels=range(corr_matrix.shape[0]))
plt.show()
```

```python
f1_exam = pd.DataFrame(columns = ["max_d","macroF1", "numFeature"])
```

```python
f1_exam
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score

# Create an instance of the LogisticRegression model
lr_model = LogisticRegression(class_weight='balanced', random_state=seed, max_iter=5000)
```

```python
# 특정 height에서 클러스터를 정의

for i in range(0, 80, 1):
    max_d = i / 10
    clusters = fcluster(linked, max_d, criterion='distance')
    
    # 각 클러스터의 대표 특성을 선택 (가장 중앙에 있는 특성 선택)
    unique_clusters = np.unique(clusters)
    selected_features_dend = []
    
    for cluster in unique_clusters:
        cluster_features = np.where(clusters == cluster)[0]
        if len(cluster_features) > 0:
            representative_feature = cluster_features[len(cluster_features) // 2]
            selected_features_dend.append(representative_feature)
    
    # 선택된 대표 특성들
    #print("Selected representative features:", selected_features_dend)
    numFeature = len(selected_features_dend)
    
    X_train_300 = X_train_scaled[:, selected_features_dend]
    X_val_300 = X_val_scaled[:, selected_features_dend]
    X_test_300 = X_test_scaled[:, selected_features_dend]

    # Train the model
    lr_model.fit(X_train_300, y_train)
    # 예측
    y_pred = lr_model.predict(X_val_300)
    f1 = f1_score(y_val, y_pred, average='macro')
    
    # Create a new row as a dictionary
    new_row = {'max_d': max_d, 'macroF1': f1, 'numFeature': numFeature}
    
    # Convert the new row to a DataFrame
    new_row_df = pd.DataFrame([new_row])
    
    # Use concat to add the new row to f1_exam
    f1_exam = pd.concat([f1_exam, new_row_df], ignore_index=True)
```

```python
# Sort f1_exam by the 'macroF1' column in descending order
f1_exam_sorted = f1_exam.sort_values(by='macroF1', ascending=False)
```
